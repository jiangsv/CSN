```python
import csv, os, random
import pandas as pd
import numpy as np
from pandas import ExcelWriter
from pandas import ExcelFile


def gen_dataframe(fleet):
    os.chdir(r"C:\Users\fx808\Desktop\exam") 
    df = pd.read_excel(f'{fleet} exam.xlsx', index_col=0).fillna('无')
# 读取excel中的题目
    _lst = []
    while len(_lst) <= 24:
        i = random.choice(range(1,(len(df.index))))
        if i not in _lst:
            _lst.append(i)
        else:
             continue
# 建立随机题库容器
    qa_df = pd.DataFrame(df, index = _lst).reset_index()
    a_s = qa_df['answer']

    line= ''
    for i in range(25):
        choiceslen = int(0)
        line = line + str(i + 1) + '：' + str(qa_df.loc[i][1]) + '\r\n'
        for n in (2, 3, 4, 5):
            choiceslen = choiceslen + len(str(qa_df.loc[i][n])) + 4 * (n - 2)
            if choiceslen<=60:
# choiceslen跟字体大小调节 小5号字体为60
                line = line + chr(95 + n).upper() + '.' + str(qa_df.loc[i][n]) + '    '
            else:
                line = line + '\r\n' + chr(95 + n).upper() + '.' + str(qa_df.loc[i][n]) + '    '
                choiceslen = int(0)
        line = line + '\r\n'
# 建立题目的字符串并排版
    _Q_outpput = (f'{fleet}   姓名______    员工号______    申请等级__    成绩____\n 选择题 每题2分（在左侧题号处写处答案)\n{line}')
    
    _answer_lst = []
    for i in range(len(a_s)):
        if i%10 == 0 :
            _answer_index = f'{i+1}-{i//10*10+10}'
            _lst.append(_answer_index)
        _answer_lst.append(a_s[i])
        
    print(_Q_outpput)    
    print(_answer_lst)

    
# 打印问题块，打印答案
def gen_sheets(*fleets):
    for i in fleets:
        gen_dataframe(i)

#  在工作目录创建文档picked_csv.csv 文档
#  xxxxxx.to_csv('picked_csv.csv', quoting=0 , sep=' ', index=False, header=False)
```
